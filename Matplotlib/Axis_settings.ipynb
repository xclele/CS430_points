```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 20)
y1 = x ** 2 + 10
y2 = x ** 3

plt.figure(num=3,figsize=(5,3),dpi=300)
plt.plot(x,y1)
plt.plot(x,y2,color = 'black',linestyle = '--',linewidth = 2)

plt.xlim((4,8)) #limit the range of the x-axis
plt.ylim((0,400)) #limit the range of the y-axis

plt.xlabel('x axis')    #Set the title of each axis
plt.ylabel('y axis')

plt.xticks([5.0],['Five point zero'])   # will replace all the original titles
plt .yticks([100,200,300,400],['One Hundred','Two Hundred','Three Hundred','Four Hundred'])

plt.show()
```

## Get the current axis

```python
x = np.linspace(0, 10, 20)
y1 = x ** 2 + 10
y2 = x ** 3

plt.figure(num=3,figsize=(5,3),dpi=300)
plt.plot(x,y1)
plt.plot(x,y2,color = 'black',linestyle = '--',linewidth = 2)

plt.xlim((4,8)) #limit the range of the x-axis
plt.ylim((0,400)) #limit the range of the y-axis

plt.xlabel('x axis')    #Set the title of each axis
plt.ylabel('y axis')

plt.xticks([5.0],['Five point zero'])   # will replace all the original titles
plt .yticks([100,200,300,400],['One Hundred','Two Hundred','Three Hundred','Four Hundred'])

ax = plt.gca()
ax.spines['top'].set_color('None')
ax.spines['right'].set_color('pink')

plt.show()
```
