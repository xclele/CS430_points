### Use figure to show seperate graphs!

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 20)
y1 = x ** 2 + 10
y2 = x ** 3
# fig1
plt.figure()
plt.plot(x,y1)
# fig2
plt.figure(num=3,figsize=(10,6))    #figsize:指定figure的宽和高，单位为英寸
plt.plot(x,y1)
plt.plot(x,y2,color = 'black',linestyle = '--',linewidth = 5)

plt.show()
```
